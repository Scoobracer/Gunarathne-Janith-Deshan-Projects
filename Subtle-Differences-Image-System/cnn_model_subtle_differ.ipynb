```python
import json
import pandas as pd

# Load JSON files into dictionaries
def load_json(file_path):
    with open(file_path, 'r') as f:
        return json.load(f)

# Load class and caption annotations
class_train = load_json('/content/dataset/class_train_annotation.json')
class_val = load_json('/content/dataset/class_val_annotation.json')
capt_train = load_json('/content/dataset/capt_train_annotations.json')
capt_val = load_json('/content/dataset/capt_val_annotations.json')
```

```python
import pandas as pd
import numpy as np
from PIL import Image
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import matplotlib.pyplot as plt

# Prepare Training Data
train_data = []
for item in class_train:
    pair_id = item['name']  # Unique identifier for each pair

    # Extract first available label
    label = item['attributes'][0]['key'] if item['attributes'] else None

    # Find corresponding caption
    caption_item = next((cap for cap in capt_train if cap['name'] == pair_id), None)
    caption = " ".join(attr['value'] for attr in caption_item['attributes']) if caption_item else None

    # Get image names
    image1_path = item['contents'][0]['name']  # First image
    image2_path = item['contents'][1]['name']  # Second image

    train_data.append({'pair_id': pair_id, 'label': label, 'caption': caption,
                       'image1': image1_path, 'image2': image2_path})

train_df = pd.DataFrame(train_data)

# Repeat for Validation Data
val_data = []
for item in class_val:
    pair_id = item['name']
    label = item['attributes'][0]['key'] if item['attributes'] else None
    caption_item = next((cap for cap in capt_val if cap['name'] == pair_id), None)
    caption = " ".join(attr['value'] for attr in caption_item['attributes']) if caption_item else None
    image1_path = item['contents'][0]['name']
    image2_path = item['contents'][1]['name']

    val_data.append({'pair_id': pair_id, 'label': label, 'caption': caption,
                     'image1': image1_path, 'image2': image2_path})

val_df = pd.DataFrame(val_data)

# Encode Labels
label_encoder = LabelEncoder()
train_labels = label_encoder.fit_transform(train_df['label'])
val_labels = label_encoder.transform(val_df['label'])

# Function to load and combine image pairs
def load_image_pair(image1_name, image2_name, img_size=(128, 128)):
    img1 = Image.open(f"/content/dataset/train/v1_1002_after.png").resize(img_size)
    img2 = Image.open(f"/content/dataset/train/v1_1002_before.png").resize(img_size)
    img1 = np.array(img1) / 255.0
    img2 = np.array(img2) / 255.0
    combined = np.concatenate([img1, img2], axis=-1)  # Shape: (128, 128, 6)
    return combined

# Load Images for Training and Validation
X_train = np.array([load_image_pair(row['image1'], row['image2']) for _, row in train_df.iterrows()])
X_val = np.array([load_image_pair(row['image1'], row['image2']) for _, row in val_df.iterrows()])

# Convert Labels to One-Hot Encoding
y_train = tf.keras.utils.to_categorical(train_labels, num_classes=3)
y_val = tf.keras.utils.to_categorical(val_labels, num_classes=3)

# Define CNN Model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 6)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(3, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Train Model
history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=64,
    validation_data=(X_val, y_val)
)

# Plot Training History
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Function to Predict Differences
def predict_difference(image1_name, image2_name):
    combined_image = load_image_pair(image1_name, image2_name)
    prediction = model.predict(np.expand_dims(combined_image, axis=0))
    predicted_class = np.argmax(prediction)
    return label_encoder.inverse_transform([predicted_class])[0]

# Example Usage
image1 = 'test_image1'
image2 = 'test_image2'
result = predict_difference(image1, image2)
print(f"Predicted difference: {result}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from PIL import Image
import json

# Function to load multiple annotation files (assuming they are in JSON format)
def load_annotations_from_multiple_files(file_paths):
    annotations = []

    for file_path in file_paths:
        with open(file_path, 'r') as f:
            data = json.load(f)
            annotations.extend(data)  # Add the contents of the current file to the annotations list

    return annotations

# Define the annotation files list BEFORE calling the function
annotation_files = [
    '/content/dataset/capt_train_annotations.json',
    '/content/dataset/capt_val_annotations.json',
    '/content/dataset/class_train_annotation.json',
    '/content/dataset/class_val_annotation.json'
]

# Load annotations from all files
annotations = load_annotations_from_multiple_files(annotation_files)

def predict_all_differences(image1_name, image2_name, annotations):
    predictions = {
        'color': 'Color: N/A',
        'shape': 'Shape: N/A',
        'texture': 'Texture: N/A',
        'image_name': image1_name
    }

    for annotation in annotations:
        for content in annotation["contents"]:
            if content["name"] == image1_name:
                for attr in annotation["attributes"]:
                    if 'value' in attr:
                        if attr["key"] == "color":
                            predictions['color'] = f"Color: {attr['value']}"
                        elif attr["key"] == "texture":
                            predictions['texture'] = f"Texture: {attr['value']}"
                        elif attr["key"] == "shape":
                            predictions['shape'] = f"Shape: {attr['value']}"

    return predictions

# Test the function
image1 = 'v3_1034_after'  # Example image name
image2 = 'v3_1034_before'

predictions = predict_all_differences(image1, image2, annotations)
print(predictions)

# Function to display prediction results
def display_full_prediction(image1_name, image2_name, annotations):
    try:
        img1 = Image.open(f"/content/dataset/train/{image2_name}.png")
        img2 = Image.open(f"/content/dataset/train/{image1_name}.png")
    except Exception as e:
        print(f"Error loading images: {e}")
        return

    predictions = predict_all_differences(image1_name, image2_name, annotations)
    print(f"Predictions: {predictions}")

    explanations = {
        "color": predictions["color"],
        "shape": predictions["shape"],
        "texture": predictions["texture"]
    }

    fig = plt.figure(figsize=(8, 6))
    gs = gridspec.GridSpec(3, 2, height_ratios=[2, 1, 1])

    ax0 = plt.subplot(gs[0, 0])
    ax0.imshow(img1)
    ax0.set_title(f"Image 1: {image1_name}")
    ax0.axis("off")

    ax1 = plt.subplot(gs[0, 1])
    ax1.imshow(img2)
    ax1.set_title(f"Image 2: {image2_name}")
    ax1.axis("off")

    ax2 = plt.subplot(gs[1:, :])
    ax2.axis("off")

    table_data = [
        ["Attribute", "Explanation"],
        ["Color", explanations["color"]],
        ["Shape", explanations["shape"]],
        ["Texture", explanations["texture"]],
    ]

    table = ax2.table(cellText=table_data, loc="center", cellLoc="center", colWidths=[0.2, 0.6],
                      cellColours=[['lightblue', 'lightblue'], ['lightyellow', 'lightyellow'], ['lightgreen', 'lightgreen'], ['lightpink', 'lightpink']])

    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1.2, 1.5)

    plt.tight_layout()
    plt.show()

# Example Usage
display_full_prediction(image1, image2, annotations)
```
